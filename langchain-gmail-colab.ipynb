````python
import os.path
import pickle
from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build

# Gmail API ìŠ¤ì½”í”„ ì„¤ì • (ì½ê¸°, ì“°ê¸°, ìˆ˜ì • ê¶Œí•œ)
SCOPES = ['https://www.googleapis.com/auth/gmail.readonly',
          'https://www.googleapis.com/auth/gmail.compose',
          'https://www.googleapis.com/auth/gmail.modify']

def get_gmail_service():
    """Gmail API ì„œë¹„ìŠ¤ ê°ì²´ë¥¼ ìƒì„±í•˜ê³  ì¸ì¦"""
    creds = None
    
    # token.pickle íŒŒì¼ì´ ìˆìœ¼ë©´ ì €ì¥ëœ ì¸ì¦ ì •ë³´ ë¡œë“œ
    if os.path.exists('token.pickle'):
        with open('token.pickle', 'rb') as token:
            creds = pickle.load(token)
    
    # ìœ íš¨í•œ ì¸ì¦ ì •ë³´ê°€ ì—†ìœ¼ë©´ ìƒˆë¡œ ë¡œê·¸ì¸
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(
                'drive/MyDrive/credentials.json', SCOPES)
            
            # Headless í™˜ê²½ìš©: redirect_urië¥¼ localhostë¡œ ì„¤ì •
            flow.redirect_uri = 'http://localhost:8080/'
            
            # ì¸ì¦ URL ìƒì„±
            auth_url, _ = flow.authorization_url(
                access_type='offline',
                include_granted_scopes='true',
                prompt='consent'
            )
            
            print("=" * 70)
            print("ğŸ” Google ê³„ì • ì¸ì¦ì´ í•„ìš”í•©ë‹ˆë‹¤ (Colab/Headless í™˜ê²½)")
            print("=" * 70)
            print("\nğŸ“‹ ë‹¤ìŒ ë‹¨ê³„ë¥¼ ë”°ë¼ì£¼ì„¸ìš”:")
            print("\n1. ì•„ë˜ URLì„ ë³µì‚¬í•˜ì—¬ ë¸Œë¼ìš°ì €ì—ì„œ ì—½ë‹ˆë‹¤:")
            print(f"\n{auth_url}\n")
            print("2. Google ê³„ì •ìœ¼ë¡œ ë¡œê·¸ì¸í•˜ê³  ê¶Œí•œì„ ìŠ¹ì¸í•©ë‹ˆë‹¤")
            print("3. ìŠ¹ì¸ í›„ ë¦¬ë””ë ‰ì…˜ëœ URL ì „ì²´ë¥¼ ë³µì‚¬í•©ë‹ˆë‹¤")
            print("   (URLì€ http://localhost:...ìœ¼ë¡œ ì‹œì‘í•˜ë©° code= íŒŒë¼ë¯¸í„°ë¥¼ í¬í•¨)")
            print("=" * 70)
            
            # ì‚¬ìš©ìë¡œë¶€í„° ë¦¬ë””ë ‰ì…˜ URL ì…ë ¥ ë°›ê¸°
            redirect_response = input("\në¦¬ë””ë ‰ì…˜ëœ URL ì „ì²´ë¥¼ ë¶™ì—¬ë„£ìœ¼ì„¸ìš”: ").strip()
            
            # URLì—ì„œ ì¸ì¦ ì½”ë“œ ì¶”ì¶œ
            code = redirect_response.split('code=')[1].split('&')[0] if 'code=' in redirect_response else redirect_response
            
            # ì¸ì¦ ì½”ë“œë¡œ í† í° êµí™˜
            flow.fetch_token(code=code)
            creds = flow.credentials
        
        # ì¸ì¦ ì •ë³´ë¥¼ token.pickleì— ì €ì¥
        with open('token.pickle', 'wb') as token:
            pickle.dump(creds, token)
    
    # Gmail API ì„œë¹„ìŠ¤ ìƒì„±
    service = build('gmail', 'v1', credentials=creds)
    return service

# Gmail ì„œë¹„ìŠ¤ ì´ˆê¸°í™”
gmail_service = get_gmail_service()
print("Gmail API ì¸ì¦ ì™„ë£Œ!")
````

### Step 2-2. Gmail ì´ë©”ì¼ ì²˜ë¦¬ í—¬í¼ í•¨ìˆ˜

````python
import base64
from email.mime.text import MIMEText

def get_unread_emails(service, max_results=5):
    """ì½ì§€ ì•Šì€ ì´ë©”ì¼ì„ ê°€ì ¸ì˜´"""
    try:
        # INBOXì˜ ì½ì§€ ì•Šì€ ì´ë©”ì¼ ê²€ìƒ‰
        results = service.users().messages().list(
            userId='me',
            labelIds=['INBOX'],
            q='is:unread',
            maxResults=max_results
        ).execute()
        
        messages = results.get('messages', [])
        
        if not messages:
            print("ì½ì§€ ì•Šì€ ì´ë©”ì¼ì´ ì—†ìŠµë‹ˆë‹¤.")
            return []
        
        email_list = []
        for message in messages:
            msg = service.users().messages().get(
                userId='me', 
                id=message['id'],
                format='full'
            ).execute()
            
            # ì´ë©”ì¼ ì •ë³´ ì¶”ì¶œ
            headers = msg['payload']['headers']
            subject = next((h['value'] for h in headers if h['name'] == 'Subject'), 'No Subject')
            sender = next((h['value'] for h in headers if h['name'] == 'From'), 'Unknown')
            
            # ì´ë©”ì¼ ë³¸ë¬¸ ì¶”ì¶œ
            body = ""
            if 'parts' in msg['payload']:
                for part in msg['payload']['parts']:
                    if part['mimeType'] == 'text/plain':
                        body = base64.urlsafe_b64decode(part['body']['data']).decode('utf-8')
                        break
            else:
                if 'body' in msg['payload'] and 'data' in msg['payload']['body']:
                    body = base64.urlsafe_b64decode(msg['payload']['body']['data']).decode('utf-8')
            
            email_list.append({
                'id': message['id'],
                'thread_id': msg['threadId'],
                'subject': subject,
                'sender': sender,
                'body': body
            })
        
        return email_list
    
    except Exception as e:
        print(f"ì´ë©”ì¼ ê°€ì ¸ì˜¤ê¸° ì˜¤ë¥˜: {e}")
        return []

def create_draft_reply(service, thread_id, to, subject, body):
    """Gmailì— ë‹µì¥ ì´ˆì•ˆì„ ìƒì„±"""
    try:
        # ì´ë©”ì¼ ë©”ì‹œì§€ ìƒì„±
        message = MIMEText(body)
        message['to'] = to
        message['subject'] = f"Re: {subject}"
        
        raw_message = base64.urlsafe_b64encode(message.as_bytes()).decode('utf-8')
        
        # Draft ìƒì„±
        draft = service.users().drafts().create(
            userId='me',
            body={
                'message': {
                    'raw': raw_message,
                    'threadId': thread_id
                }
            }
        ).execute()
        
        print(f"âœ… Draft ìƒì„± ì™„ë£Œ: {draft['id']}")
        return draft['id']
    
    except Exception as e:
        print(f"Draft ìƒì„± ì˜¤ë¥˜: {e}")
        return None

print("Gmail í—¬í¼ í•¨ìˆ˜ ì •ì˜ ì™„ë£Œ!")
````

````python
import os
from typing import TypedDict, Literal
from langgraph.graph import StateGraph, END
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate

# 1. API í‚¤ ì„¤ì • (Google Gemini API í‚¤ í•„ìš”)
os.environ["GOOGLE_API_KEY"] = "your-google-api-key"  # ì—¬ê¸°ì— Google API Key ì…ë ¥

# 2. ëª¨ë¸ ì´ˆê¸°í™” (Gemini ëª¨ë¸ ì‚¬ìš©)
llm = ChatGoogleGenerativeAI(model="gemini-flash-lite-latest", temperature=0)
````

### Step 3. LangChain ë…¸ë“œ ì •ì˜

````python
# 3. ìƒíƒœ(State) ì •ì˜: ë…¸ë“œ ê°„ì— ì£¼ê³ ë°›ì„ ë°ì´í„° êµ¬ì¡°
class EmailState(TypedDict):
    email_id: str           # Gmail ë©”ì‹œì§€ ID
    thread_id: str          # Gmail ìŠ¤ë ˆë“œ ID
    sender: str             # ë°œì‹ ì
    subject: str            # ì œëª©
    email_content: str      # ì›ë³¸ ì´ë©”ì¼ ë‚´ìš©
    category: str           # ë¶„ë¥˜ëœ ì¹´í…Œê³ ë¦¬
    draft: str              # ìƒì„±ëœ ë‹µì¥ ì´ˆì•ˆ
    draft_id: str           # Gmail Draft ID

# 4. ë…¸ë“œ í•¨ìˆ˜ ì •ì˜ (ì—ì´ì „íŠ¸ì˜ í–‰ë™)

# [Node 1] Gmailì—ì„œ ì´ë©”ì¼ ê°€ì ¸ì˜¤ê¸°
def fetch_email(state: EmailState):
    print("--- 1. Gmailì—ì„œ ì´ë©”ì¼ ê°€ì ¸ì˜¤ê¸° ---")
    # ì´ë¯¸ stateì— ì´ë©”ì¼ ì •ë³´ê°€ ìˆìœ¼ë©´ ê·¸ëŒ€ë¡œ ë°˜í™˜
    if state.get("email_content"):
        return state
    return state

# [Node 2] ì´ë©”ì¼ ë¶„ë¥˜
def classify_email(state: EmailState):
    print("--- 2. ì´ë©”ì¼ ë¶„ë¥˜ ì¤‘ ---")
    prompt = ChatPromptTemplate.from_template(
        "ë‹¤ìŒ ì´ë©”ì¼ì„ ë¶„ì„í•˜ì—¬ 'SPAM', 'REFUND', 'INQUIRY' ì¤‘ í•˜ë‚˜ë¡œ ë¶„ë¥˜í•´ì¤˜. ì˜¤ì§ ë‹¨ì–´ í•˜ë‚˜ë§Œ ì¶œë ¥í•´.\n\nì´ë©”ì¼: {email}"
    )
    chain = prompt | llm
    category = chain.invoke({"email": state["email_content"]}).content.strip()
    print(f"ë¶„ë¥˜ ê²°ê³¼: {category}")
    return {"category": category}

# [Node 3] ë‹µì¥ ì´ˆì•ˆ ì‘ì„±
def draft_reply(state: EmailState):
    print("--- 3. ë‹µì¥ ì´ˆì•ˆ ì‘ì„± ì¤‘ ---")
    category = state["category"]
    email = state["email_content"]
    
    if category == "SPAM":
        print("ìŠ¤íŒ¸ìœ¼ë¡œ ë¶„ë¥˜ë˜ì–´ ë‹µì¥ì„ ì‘ì„±í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
        return {"draft": "IGNORE"}
    
    prompt = ChatPromptTemplate.from_template(
        "ë‹¹ì‹ ì€ ì¹œì ˆí•œ ê³ ê° ì§€ì› AIì…ë‹ˆë‹¤. ë‹¤ìŒ {category} ìœ í˜•ì˜ ì´ë©”ì¼ì— ëŒ€í•œ ì •ì¤‘í•œ ë‹µì¥ì„ ì‘ì„±í•´ì£¼ì„¸ìš”.\n\nì›ë¬¸: {email}"
    )
    chain = prompt | llm
    draft = chain.invoke({"category": category, "email": email}).content
    print(f"ë‹µì¥ ì´ˆì•ˆ ìƒì„± ì™„ë£Œ (ê¸¸ì´: {len(draft)} ê¸€ì)")
    return {"draft": draft}

# [Node 4] Gmailì— Draft ì €ì¥
def save_to_gmail_draft(state: EmailState):
    print("--- 4. Gmailì— Draft ì €ì¥ ì¤‘ ---")
    
    if state["draft"] == "IGNORE":
        print("ìŠ¤íŒ¸ ë©”ì¼ì´ë¯€ë¡œ Draftë¥¼ ìƒì„±í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
        return {"draft_id": "IGNORED"}
    
    draft_id = create_draft_reply(
        service=gmail_service,
        thread_id=state["thread_id"],
        to=state["sender"],
        subject=state["subject"],
        body=state["draft"]
    )
    
    return {"draft_id": draft_id if draft_id else "FAILED"}

print("ë…¸ë“œ í•¨ìˆ˜ ì •ì˜ ì™„ë£Œ!")
````

### Step 4. LangChain Workflow êµ¬ì„± ë° ì»´íŒŒì¼

````python
# 5. ê·¸ë˜í”„(Workflow) êµ¬ì„±
workflow = StateGraph(EmailState)

# ë…¸ë“œ ì¶”ê°€
workflow.add_node("fetch", fetch_email)
workflow.add_node("classifier", classify_email)
workflow.add_node("drafter", draft_reply)
workflow.add_node("save_draft", save_to_gmail_draft)

# ì—£ì§€(íë¦„) ì—°ê²°
workflow.set_entry_point("fetch")           # ì‹œì‘ì : ì´ë©”ì¼ ê°€ì ¸ì˜¤ê¸°
workflow.add_edge("fetch", "classifier")     # ê°€ì ¸ì˜¤ê¸° -> ë¶„ë¥˜
workflow.add_edge("classifier", "drafter")   # ë¶„ë¥˜ -> ë‹µì¥ ì‘ì„±
workflow.add_edge("drafter", "save_draft")   # ë‹µì¥ ì‘ì„± -> Draft ì €ì¥
workflow.add_edge("save_draft", END)         # Draft ì €ì¥ -> ì¢…ë£Œ

# ì»´íŒŒì¼
app = workflow.compile()
print("ì›Œí¬í”Œë¡œìš° ì»´íŒŒì¼ ì™„ë£Œ!")
````
